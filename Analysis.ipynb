# Neural Network from scratch

In this assignment list we've been creating simple Neural Network from scratch to classify handwritten digits from well-known dataset MNIST.

### Architecture
Architecture of our Neural Network was quite simple. We've got 2 hidden layers with varying size depending on experiment case. For activation layer we've used Sigmoid function and for output layer Softmax.

### Training procedure:
For training procedure we've use batch size equal to $100$ with $40$ epochs where each time training set was randomly shuffled (to avoid remembering by NN). Moreover for weight updates we've used $\alpha = 0.05$.

### Experiment
We've performed experiment in which we wanted to check how does size of the model and initialization weight impacts final results. Especially we have been checkin following layers' sizes:

* (First layer size, second layer size)
* (100, 100)
* (200, 100)
* (300, 100)
* (300, 200)
* (500, 100)
* (500, 200)
* (500, 300)

and following initialization weights:

* $\mathcal{N}(0, 1)$ - Normal
* $\mathcal{N}(0, 0.1^2)$ - NormalScaled
* $\mathcal{U}(-1, 1)$ - Uniform

Results can be found in the table below.

```python
import pandas as pd
```

### Conclusions:

* The best results was obtained for the biggest Neural Network with layers' sizes (500, 300) and for Normal Scaled initialization paradigm.
* None of the models was extremly overfitted but it could be better if we would use early stopping.
* Initialization does make a difference and in this case the best one was Normal Scaled initialization, next Uniform and the worst Normal
* The differences between hidden la.yers' sizes could be seen between some results but in general was not so critical as weights initialization.
